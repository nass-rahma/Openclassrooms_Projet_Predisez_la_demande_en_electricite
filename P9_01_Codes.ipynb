# 3.1  Prediction de la consommation en un an avec la méthode Holt Winters

```python
df=dataconstem1
```

```python
df
```

```python
df.info()
```

```python
df.index.freq = 'MS'
```

```python
from statsmodels.tsa.api import ExponentialSmoothing
fit = ExponentialSmoothing(
    df,
    seasonal_periods=12,
    trend="add",
    seasonal="add",
    initialization_method="estimated",
).fit()
simulations = fit.simulate(
    12, repetitions=100, error="add", random_errors="bootstrap")

ax =df.plot(
    figsize=(10, 10),
    marker="o",
    color="black",
    title="Forecasts and simulations from Holt-Winters' multiplicative method",
)
ax.set_ylabel("Consommation")
ax.set_xlabel("Year")
fit.fittedvalues.plot(ax=ax, style="--", color="green")
simulations.plot(ax=ax, style="-", alpha=0.05, color="grey", legend=False)
fit.forecast(8).rename("Holt-Winters (add-mul-seasonal)").plot(
    ax=ax, style="--", marker="o", color="green", legend=True
)
plt.show()
```

#Mesure de Précisions sur le modèle :

```python
df['Preditesconso']=fit.fittedvalues
```

```python
df
```

```python
df['diff1']=df['Preditesconso']-df['Consommcorrig']
```

```python
mse1=((df['diff1'])**2).mean()
```

```python
print('The Mean Squared Error is {}'.format(round(mse1, 2)))
print('The Root Mean Squared Error is {}'.format(round(np.sqrt(mse1), 2)))
```

# 3.2 Prédiction de la consommation avec la méthode SARIMA

```python
df2=df[['Consommcorrig']]
```

```python
df2
```

```python
import statsmodels.api as sm
mod1=sm.tsa.statespace.SARIMAX(df2,order=(1,1,1),seasonal_order=(1,1,1,12),enforce_statonarity=False,enforce_nvertibility=False)
results1=mod1.fit()
print(results1.summary())
```

```python
results1.plot_diagnostics(figsize=(14,10))
plt.show()
```

# les paramètres du modèle P.value>0.05 : on teste un second modèle

```python
#Nous utilisons l’option , qui permet au polynôme de la moyenne mobile d’être non inversible, de sorte que plus de modèles sont estimables.enforce_invertibility=False#
import statsmodels.api as sm
mod=sm.tsa.statespace.SARIMAX(df2,order=(1,1,0),seasonal_order=(1,0,1,12),enforce_statonarity=False,enforce_nvertibility=False)
results=mod.fit()
print(results.summary())
```

```python
# modèle 
```

```python
results.plot_diagnostics(figsize=(14,10))
plt.show()
```

# Les paramètres du modèle P.value <0.05 et le bruit blanc >0.05: modèle validé

```python
# prédictions et interval de confiance
#12etapes
forecast_values=results.get_forecast(steps=12)
forecast_ci=forecast_values.conf_int()
```

```python
forecast_values.summary_frame()
```

```python
# graphique 
ax=df2.plot(figsize=(12,12))
forecast_values.predicted_mean.plot(ax=ax,label='Forecasts')
ax.fill_between(forecast_ci.index,forecast_ci.iloc[:,0],forecast_ci.iloc[:,1],color='g',alpha=.5)

ax.set_xlabel('Mois')
ax.set_ylabel('Consommation')
ax.set_title('SARIMA')
plt.legend()
```

#Mesure de Précisions sur le modèle :

```python
pred = results.get_prediction(start=pd.to_datetime('2020-08-01'), dynamic=False)
pred_ci = pred.conf_int()
ax = df2['2019':].plot(label='observed')
pred.predicted_mean.plot(ax=ax, label='forecasts', alpha=.7, figsize=(14, 4))
ax.fill_between(pred_ci.index,
                pred_ci.iloc[:, 0],
                pred_ci.iloc[:, 1], color='k', alpha=.2)
ax.set_xlabel('Mois')
ax.set_ylabel('Consommation')
plt.legend()
plt.show()
```
